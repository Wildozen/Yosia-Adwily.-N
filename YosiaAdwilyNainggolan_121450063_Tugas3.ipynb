```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Menggunakan dataset iris dari seaborn
iris = sns.load_dataset("iris")

# Melihat 5 baris pertama dari dataset
print(iris.head())
```

```python
# Membuat histogram untuk melihat distribusi panjang dan lebar kelopak/mahkota
sns.set(style="whitegrid")
iris.hist(figsize=(10, 8), bins=20)
plt.suptitle('Distribusi Fitur Dataset Iris', fontsize=16)
plt.show()
```

```python
# Membuat scatter plot untuk menggambarkan hubungan antara fitur berdasarkan spesies
sns.pairplot(iris, hue="species", markers=["o", "s", "D"], height=2.5)
plt.suptitle('Hubungan Antar Fitur Dataset Iris', fontsize=16)
plt.show()
```

```python
# Membuat boxplot untuk menunjukkan distribusi fitur berdasarkan spesies
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title('Distribusi Panjang Kelopak Berdasarkan Spesies', fontsize=14)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title('Distribusi Panjang Mahkota Berdasarkan Spesies', fontsize=14)
plt.show()
```
